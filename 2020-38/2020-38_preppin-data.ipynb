## Test if words are anagram

- Anagrams are formed by re-arranging of another word (on the same row)
- All anagrams are one word only
- No letter can be used more than once
- All letters must be used

```python
# Solution 1: List comprehension
df["Anagram?"] = ["Yes" if sorted(df["Word 1"][i].lower()) == sorted(df["Word 2"][i].lower()) else "No" for i in range(len(df))]
df
```

```python
# Solution 2: For loop
anagram = []

for i in range(len(df)):
    if sorted(df["Word 1"][i].lower()) == sorted(df["Word 2"][i].lower()):
        test = "Yes"
    else:
        test = "No"
    anagram.append(test)
    
df["Anagram?"] = anagram
df
```
